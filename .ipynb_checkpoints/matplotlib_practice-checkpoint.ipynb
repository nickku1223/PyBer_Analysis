```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Line Chart

#### MATLAB method

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

#### Object-oriented interface

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

##### Annotation with MATLAB interface

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

##### Skill drill (annotation with object-oriented interface)

```python
# Change the line color to green
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker = "D", color = "green", linewidth = 2, label = "Boston")

# Add legend for the city of Boston
ax.legend()

# Set title and labels for x-axis and y-axis
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set y-axis limit
ax.set_ylim(0, 45)

# Add grid lines
ax.grid()
```

## Bar Chart

MATLAB approach

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color = "green", label = "Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a titile
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

#### Horizontal bar chart

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot (opposite axes)
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
#### Skill drill (changing the graph with MATLAB approach)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.title("PyBer Fare by Month")
plt.legend()
```

Object-orientec approach

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (horizontal bar)
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt (opposite axes)
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

Skill drill

```python
# Change the bar color to cyan.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Invert the y-axis data.

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = "Chicago")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
```

# Scatter plot

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

Skill Drill (using the MATLAB plotting approach)

```python
# Change the color of the markers to red.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

plt.scatter(y_axis, x_axis, c = "red", label = "Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.legend()
plt.xlim(0, 45)
plt.gca().invert_yaxis()
```
