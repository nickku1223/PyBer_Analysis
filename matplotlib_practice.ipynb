```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Line Chart

#### MATLAB method

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

#### Object-oriented interface

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

##### Annotation with MATLAB interface

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

##### Skill drill (annotation with object-oriented interface)

```python
# Change the line color to green
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker = "D", color = "green", linewidth = 2, label = "Boston")

# Add legend for the city of Boston
ax.legend()

# Set title and labels for x-axis and y-axis
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Set y-axis limit
ax.set_ylim(0, 45)

# Add grid lines
ax.grid()
```

Skill Drill (using the MATLAB plotting approach)

```python
# Change the color of the markers to red.
# Add a legend for the city of Chicago.
# Add a title and axes labels.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

plt.scatter(y_axis, x_axis, c = "red", label = "Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Month")
plt.legend()
plt.xlim(0, 45)
plt.gca().invert_yaxis()
```

# Bubble Chart

```python
plt.scatter(x_axis, y_axis, s = y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# Use list comprehension to replace for and while loops
# format for comprehension: new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

Scatter plot with object-oriented interface:

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# To create bubble chart
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

Skill drill:

```python
# Change the color of the markers to sky blue.
# Change the size of the markers to 5 times each data point.
# Make the color 20% transparent.
# Add a black edge color to the circles.
# Make the linewidth of the circles 2 points.
# Add a legend for the city of Boston.
# Add a title.
# Switch the axis so that the Fare($) data is on the x-axis.
# Add a limit for the x-axis data.
# Invert the y-axis so that January is at the top.

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="xkcd:sky blue", s = [i * 5 for i in y_axis], alpha = 0.8, edgecolor = "black", lw=2, label = "Boston")
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_xlim(0, 45)
ax.invert_yaxis()
```

# Pie Chart

MATLAB method:

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0, )
plt.pie(y_axis, explode = explode_values, labels = x_axis, autopct = "%.1f%%")
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

Object-oriented interface:

```python
fig, ax = plt.subplots()
ax.pie(y_axis, labels = x_axis)
plt.show()
```

Skill drill:

```python
# Add a percentage to one decimal place to each wedge of the pie.
# Increase the figure size to 8x8.
# Explode the two highest percentage months.
# Add a shadow.
# Add a start angle so that January is at the top.
# Reverse the order so that the month order is in a clockwise direction.
# Add new colors of your own choosing or use the colors from the previous pie chart.


explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize = (8, 8))
ax.pie(y_axis, 
       labels = x_axis,
       explode = explode_values,
       autopct = "%.1f%%",
       startangle = 270,
       shadow = True,
       colors = colors)
ax.invert_yaxis()
plt.show()
```
